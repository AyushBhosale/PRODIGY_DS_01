```python
sns.countplot(data=df,x='Type')
```

From above graph we now know out of all the excercised running is widely implemented
Now lets find best excercise for burning callorie


```python
df.groupby(by='Type')['Calories Burned'].mean().index
```

```python
plt.bar(df.groupby(by='Type')['Calories Burned'].mean().index,df.groupby(by='Type')['Calories Burned'].mean())
plt.ylabel("Calorie Burned")
plt.xlabel("Excercises")
plt.show()
```

So By walking we are burning more calorie
But why is it so because of walking for long distances or something else


```python
df.groupby(by='Type')['Distance (km)'].mean()
plt.bar(df.groupby(by='Type')['Distance (km)'].mean().index,df.groupby(by='Type')['Distance (km)'].mean())
plt.ylabel("Distance")
plt.xlabel("Excercise")
plt.show()
```
